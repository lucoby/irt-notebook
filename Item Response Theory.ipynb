```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize as op
import seaborn as sns
sns.set_theme()
```

```python
def func_3pl(x, a, b, c):
    return (1 - c) / (1 + np.exp(-a * (x - b)))
```

```python
def test_func(x):
    return func_3pl(x, 8, .5, .25)
```

```python
x_actual = np.linspace(0, 1)
y_actual = test_func(x_actual)
```

```python
plt.plot(x_actual, y_actual, "-b")
```

```python
samples = 100
np.random.seed(seed=42)
x_test = np.random.rand(samples)
z_test = np.random.rand(samples)
y_test = z_test < test_func(x_test)
```

```python
plt.plot(x_test, y_test, ".b")
```

```python
p, _ = op.curve_fit(func_3pl, x_test, y_test)
```

```python
x_fit = np.linspace(0, 1)
y_fit = func_3pl(x_fit, p[0], p[1], p[2])
```

```python
plt.plot(x_test, y_test, ".b", x_actual, y_actual, "-g", x_fit, y_fit, "-r")
```
